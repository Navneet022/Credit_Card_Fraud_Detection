Importing the libraries


```python
import numpy as npy
import pandas as pnd
from sklearn.model_selection import train_test_split as split_data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score as calc_accuracy
from sklearn.preprocessing import StandardScaler
```

```python
# Importing the dataset into a DataFrame using pandas
data_credit_cards = pnd.read_csv('/content/creditcard.csv')
```

```python
# Displaying the initial 5 entries of the dataset
data_credit_cards.head()
```

```python
# Displaying the last 5 entries of the dataset
data_credit_cards.tail()
```

```python
# Information about the dataset
data_credit_cards.info()
```

```python
# Checking for missing values in each column
missing_values = data_credit_cards.isnull().sum()
```

```python
# Distribution of legitimate and fraudulent transactions
transaction_distribution = data_credit_cards['Class'].value_counts()
```

This Dataset is highly unblanced

0 --> Normal Transaction

1 --> fraudulent transaction


```python
# Separating the data for analysis
legitimate_transactions = data_credit_cards[data_credit_cards['Class'] == 0]
fraudulent_transactions = data_credit_cards[data_credit_cards['Class'] == 1]
```

```python
# Displaying the shape of legitimate and fraudulent transactions
print("Legitimate transactions shape:", legitimate_transactions.shape)
print("Fraudulent transactions shape:", fraudulent_transactions.shape)
```

```python
# Statistical measures of the legitimate transactions
legitimate_amount_stats = legitimate_transactions['Amount'].describe()
print(legitimate_amount_stats)
```

```python
# Statistical measures of the fraudulent transactions
fraudulent_amount_stats = fraudulent_transactions['Amount'].describe()
print(fraudulent_amount_stats)
```

```python
# Compare the average values for both legitimate and fraudulent transactions
average_values_by_class = data_credit_cards.groupby('Class').mean()
print(average_values_by_class)
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
# Randomly sample 492 entries from the legitimate transactions
sampled_legitimate = legitimate_transactions.sample(n=492)
```

Joining two Data Frames

```python
# Combine the sampled legitimate transactions with the fraudulent transactions
combined_dataset = pnd.concat([sampled_legitimate, fraudulent_transactions], axis=0)
```

```python
# Display the first few rows of the combined dataset
combined_dataset_preview = combined_dataset.head()
print(combined_dataset_preview)
```

```python
# Display the last few rows of the combined dataset
combined_dataset_summary = combined_dataset.tail()
print(combined_dataset_summary)
```

```python
# Count the occurrences of each class in the combined dataset
class_distribution = combined_dataset['Class'].value_counts()
print(class_distribution)
```

```python
# Calculate the mean of each numerical column grouped by 'Class' in the dataset
mean_by_class = combined_dataset.groupby('Class').mean()
print(mean_by_class)
```

Splitting the data into features & targets

```python
# Separate features and target variable from the dataset
features = combined_dataset.drop(columns='Class', axis=1)
target = combined_dataset['Class']
```

```python
# Display the features DataFrame
print(features)
```

```python
# Display the target variable Series
print(target)
```

Spliting the data into Training and testing data

```python
# Split the dataset into training and testing sets
features_train, features_test, target_train, target_test = split_data(
    features, target, test_size=0.2, stratify=target, random_state=2)
```

```python
# Display the shapes of the original and split feature datasets
print(features.shape, features_train.shape, features_test.shape)
```

Training the Model

Logistic Regression

```python
# Scale the features
scaler = StandardScaler()
features_train = scaler.fit_transform(features_train)
features_test = scaler.transform(features_test)

# Train the logistic regression model with increased max_iter
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(features_train, target_train)
```

Model Evaluation

Accuracy Score


```python
# Calculate accuracy on the training data
train_predictions = logistic_model.predict(features_train)
training_accuracy = calc_accuracy(train_predictions , target_train)
```

```python
# Print the accuracy on the training data
print('Accuracy on Training Data:', training_accuracy)
```

```python
# Calculate accuracy on the test data
test_predictions = logistic_model.predict(features_test)
test_accuracy = calc_accuracy(target_test, test_predictions)
```

```python
# Print the accuracy score on the test data
print('Accuracy Score on Test Data:', test_accuracy)
```
